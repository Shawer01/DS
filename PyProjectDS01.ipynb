```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tesla_data = yf.Ticker("TSLA")
```

```python
tesla_data = tesla_data.history(period="max")
```

```python
tesla_data.reset_index(inplace=True)
tesla_data.to_csv('tesla_data.csv', index=False)
```

```python
print(tesla_data.head())
```

```python
tesla_graph = tesla_data.plot(x="Date", y="Low")
plt.title("TESLA Date VS Low")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib_inline
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
```

```python
driver.get(url)
time.sleep(5)
data = driver.page_source
driver.quit()
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
tesla_revenue = pd.DataFrame(columns=["Year","Revenue"])
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    datea = col[1].text
    
    tesla_revenue = pd.concat([tesla_revenue,pd.DataFrame({"Year":[date], "Revenue":[datea]})])    
```

```python
tesla_revenue.tail()
```
